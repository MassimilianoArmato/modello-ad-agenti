```python
def divide_number(n, N):
    # Genera N-1 punti di divisione casuali
    points = sorted(random.sample(range(1, n), N - 1))
    # Aggiungi i punti di inizio e fine
    points = [0] + points + [n]
    # Calcola le differenze tra i punti per ottenere le parti
    parts = [points[i+1] - points[i] for i in range(len(points) - 1)]
    return parts

# Esempio di utilizzo
n = 20
N = 5
parts = divide_number(n, N)
print(parts)
```

```python
def process_vector(vector, parts):
    start = 0
    segments=[]
    for part in parts:
        end = start + part
        segment = vector[start:end]
        # Qui puoi fare qualcosa con il segmento, ad esempio stamparlo
        #print(segment)
        start = end
        segments.append(segment)
    return segments

# Esempio di utilizzo
vector = ['a','b','c','d','e','f','g','h','i','l'] # Un vettore di esempio con 20 elementi
parts = [1, 2, 3, 1, 3]  # La lista delle parti

process_vector(vector, parts)
```

```python
def divide_list(lst, n):
    # Calcola la dimensione di ogni parte
    k, m = divmod(len(lst), n)
    return [lst[i * k + min(i, m):(i + 1) * k + min(i + 1, m)] for i in range(n)]

# Esempio di utilizzo
vettore = list(range(1, 20))  # Vettore da dividere
parti = 10
diviso = divide_list(vettore, parti)
print(diviso)  # Output: [[1, 2], [3, 4], [5, 6]]
```

```python
def aggiungi_o_sostituisci(lista_di_liste, nuova_lista):
    for i, lista in enumerate(lista_di_liste):
        if any(nuova_lista[j] == lista[j] for j in range(len(nuova_lista))):
            lista_di_liste[i] = nuova_lista
            return
    lista_di_liste.append(nuova_lista)

# Esempio di utilizzo
lista_di_liste = [[1, 2], [3, 4], [5, 6]]
nuova_lista = [3,7]

aggiungi_o_sostituisci(lista_di_liste, nuova_lista)
print(lista_di_liste)
```

```python
def agg_or_sos2(lista_di_liste, nuova_lista):
    for i, lista in enumerate(lista_di_liste):
        if any(nuova_lista[1]==lista[1] and nuova_lista[2]>=lista[2] for j in range(len(nuova_lista))):
            lista_di_liste[i] = nuova_lista
            return
        elif any(nuova_lista[1]==lista[1] and nuova_lista[2]<lista[2] for j in range(len(nuova_lista))) :
            return
    lista_di_liste.append(nuova_lista)

lista_di_liste = [[1, 2,3], [3, 4,5], [5, 6,8]]
nuova_lista = [7,5,2]

agg_or_sos2(lista_di_liste, nuova_lista)
print(lista_di_liste)
```

```python

class Saver(Agent):
   def __init__(self, unique_id, model, schedule_savers):
        super().__init__(unique_id, model)
        self.schedule_savers = schedule_savers
        self.bank_account = random.randint(10000, 100000)
        self.banca_di_appartenenza=0
        self.conoscenza=0
        self.marker=-1
        self.info_var_equity=[]
        self.info_var_clienti=[]


class Loan(Agent):
   def __init__(self, unique_id, model, schedule_loans):
        super().__init__(unique_id, model)
        self.schedule_loans = schedule_loans
        self.random_request = 0
        self.prob_default=random.uniform(5,10)
        #print(f"loan_prob_default={self.prob_default:.2f}")
        self.loan_interest=0

class Bank(Agent):
   def __init__(self, unique_id, model, schedule_banks):
        super().__init__(unique_id, model)
        self.schedule_banks = schedule_banks
        self.bank_equity = random.randint(1e5, 2e5)
        #self.bank_equity =0
        self.cash_depositi=0
        self.numero_clienti=0
        self.bank_regione=0
        self.bank_loans=0
        self.old_equity=0
        self.info_var_equity=[]
        self.info_var_clienti=[]
        self.bank_request=0
        self.bank_default = 0





class Central_Bank(Agent):
   def __init__(self, unique_id, model, schedule_central_bank):
        super().__init__(unique_id, model)
        self.schedule_central_bank = schedule_central_bank
        self.info_numero_clienti=0

   def default(self,bank):
    for regione in model.regioni:
      if regione.id==bank.bank_regione:
        k=len(regione.saver_per_regione)
        banche_nella_regione=len(regione.banks_per_regione)
    K= self.info_numero_clienti/k
    print(f"numero clienti regione= {k}")
    print(f"rapporto={K}")
    if K<0.50 and banche_nella_regione!=1:
          print(f"\n")
          print(f"Bank {bank.unique_id} è stata rimossa")
          print(f"\n")
          bank.bank_default = 1
          return 0
    else:
      bank.bank_equity=0.5*bank.old_equity
      print(f"banca {bank.unique_id} si salva")
      return 1






class Regione():
  def __init__(self,id):
    self.id=id
    self.saver_per_regione=[]
    self.loans_per_regione=[]
    self.banks_per_regione=[]

class Blog():
  def __init__(self,model):
    self.list_var_equity=[]
    self.list_var_clienti=[]
    self.info_regioni_clienti={}
    self.info_regioni={}
    for i, regione in enumerate(model.regioni):
      self.info_regioni[f'info_var_equity_regione{i}']=[]
      self.info_regioni_clienti[f'info_var_clienti_regione{i}']=[]

  def numero_ciclo(self,i):
    self.ciclo=i


  def receive_info_clienti(self, info2):
    agg_or_sos2(self.list_var_clienti,info2)
    for i, regione in enumerate(model.regioni):
      for bank in regione.banks_per_regione:
        if bank.unique_id==info2[1]:
          agg_or_sos2(self.info_regioni_clienti[f'info_var_clienti_regione{i}'],info2)

  def receive_info(self, info):
    agg_or_sos2(self.list_var_equity,info)
    for i, regione in enumerate(model.regioni):
      for bank in regione.banks_per_regione:
        if bank.unique_id==info[1]:
          agg_or_sos2(self.info_regioni[f'info_var_equity_regione{i}'],info)



  def send_info_bc(self,bank,bc):
    bc.info_numero_clienti=bank.numero_clienti
    print(f"numero clienti = {bc.info_numero_clienti} della banca {bank.unique_id}")

  def send_info_onniscente(self):
    print(f"list_var_equity = {self.list_var_equity}")
    #print(f"info_regioni = {self.info_regioni}")
    for bank in model.schedule_banks.agents:
      bank.info_var_equity= self.list_var_equity
      bank.info_var_clienti= self.list_var_clienti
      #print(f"bank_info_var_equity {bank.unique_id}= {bank.info_var_equity}")

  def send_info_local(self):
    for i, regione in enumerate(model.regioni):
      #n=random.randint(1,len(regione.saver_per_regione))
      n = 0.5*len(regione.saver_per_regione)
      n=int(n)
      savers_scelti= random.sample(regione.saver_per_regione,n)
      for saver in savers_scelti:
        #print(f"saver_scelto in regione {regione.id} = {saver.unique_id}")
        if self.list_var_equity!=[]:
          info= random.choice(self.info_regioni[f'info_var_equity_regione{i}'])
          #print(f"info = {info} per saver {saver.unique_id}")
          agg_or_sos2(saver.info_var_equity,info)
          for bank in regione.banks_per_regione:
            if bank.unique_id==info[1]:
              for l in saver.Trust_banks:
                if l[1]==bank.unique_id:
                  T=0.5*(l[0]+sigmoid(info[0]))
                  a = [T,bank.unique_id,self.ciclo]
                  agg_or_sos2(saver.Trust_banks, a)
        if self.list_var_clienti!=[]:
          info2= random.choice(self.info_regioni_clienti[f'info_var_clienti_regione{i}'])
          #print(f"info = {info2} per saver {saver.unique_id}")
          agg_or_sos2(saver.info_var_clienti,info2)
          for bank in regione.banks_per_regione:
            if bank.unique_id==info2[1]:
              for l in saver.Trust_banks:
                if l[1]==bank.unique_id:
                  T=0.5*(l[0]+sigmoid(info2[0]))
                  a = [T,bank.unique_id,self.ciclo]
                  agg_or_sos2(saver.Trust_banks, a)


class Network(Model):

      def __init__(self, N_savers, N_loans, N_banks,N_regions):                                  ######## classe del modello: parto inizializzando i savers vari: simultaneous serve per far partire le cose una dopo l altra e non a caso
        super().__init__()
        #self.schedule_savers = RandomActivation(self)
        self.schedule_savers = SimultaneousActivation(self)
        self.schedule_loans = SimultaneousActivation(self)
        self.schedule_banks = SimultaneousActivation(self)
        self.schedule_central_bank = SimultaneousActivation(self)


        self.list_savers= []
        self.all_savers_id= []




        # Create saver agents
        for i in range(N_savers):                     ##### creazione di ogni savers con add. savers posizionati a caso nella griglia
            a = Saver(i, self, self.schedule_savers)
            self.list_savers.append(a)


        self.all_loans_id=[]
        self.list_loans=[]

        # Create loan agents
        for i in range(N_loans):                        ##### creazione di ogni loans con add
            a = Loan(i , self, self.schedule_loans)  # Continue the unique_id count
            self.list_loans.append(a)



        print(f"\n")

        self.all_banks_id=[]

        # Create bank agents
        for i in range(N_banks):
            a = Bank(i , self, self.schedule_banks)  # Continue the unique_id count                 ##### creazione di ogni banks con add. banks posizionati a caso nella griglia
            self.schedule_banks.add(a)
            self.all_banks_id.append(i)


        #create central_bank
        bc=Central_Bank(0,self,self.schedule_banks)
        self.schedule_central_bank.add(bc)

        #create blog

        print(f"\n")




#creazione categorie e depositi savers
        num_excluded = int(0.23 * N_savers)                                               ##### qui sto distinguendo savers in 4 categorie esclusi esperti ecc ecc
        num_incompetent = int(0.33 * N_savers)
        num_competent = int(0.26 * N_savers)

        # Assegnazione delle categorie ai saver
        for i, saver in enumerate(self.list_savers):
            if i < num_excluded:
                saver.category = "Esclusi"
                saver.conoscenza= random.uniform(0, 0.3)
                saver.carisma= random.uniform(0, 1)
                saver.peso_conoscenza_interlocutore = saver.conoscenza
                saver.peso_carisma_interlocutore = 1-saver.conoscenza

            elif i < num_excluded + num_incompetent:
                saver.category = "Incompetenti"
                saver.conoscenza= random.uniform( 0.3,0.6)
                saver.carisma= random.uniform(0, 1)
                saver.peso_conoscenza_interlocutore = saver.conoscenza
                saver.peso_carisma_interlocutore = 1-saver.conoscenza

            elif i < num_excluded + num_incompetent + num_competent:
                saver.category = "Competenti"
                saver.conoscenza= random.uniform(0.6,0.8)
                saver.carisma= random.uniform(0, 1)
                saver.peso_conoscenza_interlocutore = saver.conoscenza
                saver.peso_carisma_interlocutore = 1-saver.conoscenza

            else:
                saver.category = "Esperti"
                saver.conoscenza= random.uniform(0.8,1)
                saver.carisma= random.uniform(0, 1)
                saver.peso_conoscenza_interlocutore = saver.conoscenza
                saver.peso_carisma_interlocutore = 1-saver.conoscenza


        random.shuffle(self.list_savers)
        for i, saver in enumerate(self.list_savers):
            self.schedule_savers.add(saver)




        for i, saver in enumerate(self.schedule_savers.agents):
             print(f"Saver {saver.unique_id} {saver.category}  deposito = {saver.bank_account:.2f}")
             print(f"Saver {saver.unique_id} {saver.category}  conoscenza = {saver.conoscenza:.2f}")
             self.all_savers_id.append(i)
        #print(self.all_savers_id)

        print(f"\n")



        print(f"\n")
# Creazione loans e classificazione

        num_benestanti = int(0.70 * N_loans)
        loans_ricchi=[]
        loans_benestanti=[]


        for i, loans in enumerate(self.list_loans):
            if i < num_benestanti:
                loans.category = "benestante"
                loans.loan_request= random.randint(10000, 100000)

            else:
                loans.category = "ricco"
                loans.loan_request= random.randint(100000, 1000000)
        random.shuffle(self.list_loans)

        for i, loans in enumerate(self.list_loans):
            self.schedule_loans.add(loans)
            self.all_loans_id.append(i)

       # for loan in self.schedule_loans.agents:
           # print(f"Loan {loan.unique_id} {loan.category}")

        print(f"\n")

#inizializzazione banche
        def sum_savers_account(self, agent_ids):
              return sum(self.schedule_savers.agents[i].bank_account for i in agent_ids)


        savers_in_banks=divide_number(N_savers, N_banks) #divido i savers in un numero casuale per ogni banca
        print(savers_in_banks)
        print(f"\n")
        self.segments= process_vector(self.all_savers_id, savers_in_banks) #creo delle liste una per banca con dentro i savers della banca
        numero_clienti_banche=[]
        depositi=[]
        print(self.segments)
        for segment in self.segments:
            #print(segment)
            numero_clienti_banche.append(len(segment))
            depositi.append(sum_savers_account(self,segment))
       # print(depositi)
        print(f"\n")
        #print(numero_clienti_banche)
        print(f"\n")

        for i, bank in enumerate(self.schedule_banks.agents):
            bank.numero_clienti=numero_clienti_banche[i]
            bank.cash_depositi=depositi[i]
            print(f"Bank {bank.unique_id} numero clienti = {bank.numero_clienti} e deposito= {bank.cash_depositi}") #associo ad ogni banca il numero dei clienti

        print(f"\n")

#regioni banca
        lista_regioni_banks= divide_list(self.all_banks_id, N_regions)
        for i, regione in enumerate(lista_regioni_banks):
         for j in regione:
            self.schedule_banks.agents[j].bank_regione=i
            print(f"regione_banca {j} = {self.schedule_banks.agents[j].bank_regione}")
        print(f"\n")
#regioni loans
        lista_regioni_loans= divide_list(self.all_loans_id, N_regions)
        for i, regione in enumerate(lista_regioni_loans):
         for j in regione:
            self.schedule_loans.agents[j].loans_regione=i
            #print(f"regione_loans {self.schedule_loans.agents[j].unique_id} = {self.schedule_loans.agents[j].loans_regione}")
        print(f"\n")




#savers per banca
        for i,saver in enumerate(self.segments):
          for j in saver:
            self.schedule_savers.agents[j].banca_di_appartenenza=i
            self.schedule_savers.agents[j].saver_regione=self.schedule_banks.agents[i].bank_regione
            print(f"saver{self.schedule_savers.agents[j].unique_id}_banca_di_appartenenza = {self.schedule_savers.agents[j].banca_di_appartenenza}")
            print(f"saver{self.schedule_savers.agents[j].unique_id}_regione = {self.schedule_savers.agents[j].saver_regione}")
            print(f"saver{self.schedule_savers.agents[j].unique_id}_category= {self.schedule_savers.agents[j].category}")
        print(f"\n")





#regioni savers


#inizializzo trust
        for i, saver in enumerate(self.schedule_savers.agents):
              saver.Trust=[]
              saver.Trust_banks=[]
              for j, bank in enumerate(self.schedule_banks.agents):
                if saver.saver_regione==bank.bank_regione:
                  l=random.uniform(-0.010,0.010)
                  l=float(f"{l:.3f}")
                  saver.Trust_banks.append([0.500+l,bank.unique_id,0])
              print(f"Trust_banks saver {saver.unique_id} = {saver.Trust_banks}")


#creazione vettori regione

        self.regioni=[]
        for i , regione in enumerate(lista_regioni_banks):
          a=Regione(i)
          self.regioni.append(a)

        for i, regione in enumerate(self.regioni):
          for j, saver in enumerate(self.schedule_savers.agents):
            if saver.saver_regione==i:
              regione.saver_per_regione.append(saver)
          for z, bank in enumerate(self.schedule_banks.agents):
            if bank.bank_regione==i:
              regione.banks_per_regione.append(bank)

          for k, loan in enumerate(self.schedule_loans.agents):
            if loan.loans_regione==i:
              regione.loans_per_regione.append(loan)
          #for i, saver in enumerate(regione.saver_per_regione):
           # print(f"saver {saver.unique_id} appartiene a {regione.id}")
          #print(f"\n")

          self.blog=Blog(self)

          print(f"\n")



          self.banks_to_remove=[]

#---------------------------- Creazione del ciclo----------------------------------------

      def step(self,fl,ciclo):


        #loan chiedono prestiti
          for loans in self.schedule_loans.agents:
              if loans.category == "benestante":
                  loans.loan_request= random.randint(10000, 100000)
              else:
                  loans.loan_request= random.randint(100000, 1000000)
         # for loan in self.schedule_loans.agents:
            #print(f"Loan {loan.unique_id} {loan.category} richiesta di prestito = {loan.loan_request:.2f}")


          print(f"\n")

          for bank in self.schedule_banks.agents:
            bank.bank_loans=0


          #si calcolano interessi vengono decise le aste e vengono ridati i soldi
          for loan in self.schedule_loans.agents:
              list_interest=[]
              for j, bank in enumerate(self.schedule_banks.agents):
                  if loan.loans_regione==bank.bank_regione:
                    if bank.bank_loans + loan.loan_request < bank.bank_equity*10 and bank.bank_loans + loan.loan_request<bank.cash_depositi*0.9:
                      interest= 0.01+ loan.prob_default/100+random.uniform(-3,3)/100
                      list_interest.append([interest, bank.unique_id])
              #print(f"loan {loan.unique_id} interest= {list_interest}")
              if list_interest!=[] and len(list_interest)>1:
                  loan.loan_interest=min(list_interest)
                  for bank in self.schedule_banks.agents:
                    if loan.loan_interest[1]==bank.unique_id:
                      bank.bank_loans += loan.loan_request

                  #print(f"loan {loan.unique_id} interest= {loan.loan_interest}")
              else:
                loan.loan_interest= None

          for bank in self.schedule_banks.agents:
            #print(f"Bank {bank.unique_id} equity= {bank.bank_equity}")
            #print(f"Bank {bank.unique_id} loans= {bank.bank_loans}")
            bank.old_equity=bank.bank_equity
          print(f"\n")




        #fluttuazioni depositi savers a causa di stipendi e lavori
          for saver in self.schedule_savers.agents:
            fluttuazione=  random.uniform(-fl, fl)
            new_balance = saver.bank_account * (1 + fluttuazione)
            saver.bank_account = max(new_balance, 0)
           # print(f"Saver {saver.unique_id} {saver.category}  deposito = {saver.bank_account:.2f}")

          print(f"\n")
        #regioni saver
         # for saver in self.schedule_savers.agents:
          #  print(f"saver{saver.unique_id}_regione = {saver.saver_regione}")
          print(f"\n")


         #savers spese impreviste grosse
          for saver in self.schedule_savers.agents:
            prob_spesa = random.uniform(0,1)
            if prob_spesa<0.10:
              spesa_imprevista_grossa= saver.bank_account*0.5
              saver.bank_account= saver.bank_account-spesa_imprevista_grossa
             # print(f"Saver {saver.unique_id} {saver.category}  deposito = {saver.bank_account:.2f}")

          print(f"\n")



        #ricerca interlocutore savers

          def incontro(saver, depth=0):
              if depth > len(self.schedule_savers.agents):  # Limite di profondità per evitare ricorsione infinita
                  return
              if saver.marker != -1:
                  return
              saver_interlocutore = random.choice(self.schedule_savers.agents)
              if saver_interlocutore.marker == -1 and saver_interlocutore.unique_id != saver.unique_id and saver.marker==-1 and saver_interlocutore.saver_regione==saver.saver_regione: #aggiungendo il vincolo di saver.saver_regione==saver_interlocutore.saver_regione forse dovrei risolvere il problema della divisione in regioni
                  saver.marker = saver_interlocutore.unique_id
                  saver_interlocutore.marker = saver.marker
                  #print(f"Interazione tra saver {saver.unique_id} e saver {saver_interlocutore.unique_id}")
              elif saver_interlocutore.unique_id == saver.unique_id:
                  incontro(saver, depth + 1)
              elif saver_interlocutore.marker != -1:
                  incontro(saver, depth + 1)
              elif saver_interlocutore.saver_regione!=saver.saver_regione:
                  incontro(saver, depth + 1)

          # savers comunicano info tra loro
          for i, saver in enumerate(self.schedule_savers.agents):
              incontro(saver)
          print(f"\n")

          for i, saver in enumerate(self.schedule_savers.agents):
              if saver.marker==-1:
                print(f"saver {saver.unique_id} non ha trovato un interlocutore")
          print(f"\n")


         # for i, saver in enumerate(self.schedule_savers.agents):
        #     print(f"saver{saver.unique_id} ha marker= {saver.marker}")
          print(f"\n")

          for saver in self.schedule_savers.agents:
              for saver_interlocutore in self.schedule_savers.agents:
                if saver_interlocutore.marker == saver.marker and saver_interlocutore.unique_id!= saver.unique_id and saver.marker!=-1 and saver_interlocutore.marker!=-1 :
                  #print(f"comunicazione tra saver {saver.unique_id} di banca= {saver.banca_di_appartenenza} e saver {saver_interlocutore.unique_id} di banca {saver_interlocutore.banca_di_appartenenza}")
                  T= saver.peso_conoscenza_interlocutore * saver_interlocutore.conoscenza + saver.peso_carisma_interlocutore * saver_interlocutore.carisma      #trust saver vs saver
                  T=float(f"{T:.3f}")
                  a= [T,saver_interlocutore.unique_id,ciclo]
                  agg_or_sos2(saver.Trust, a)
                  for k in saver_interlocutore.Trust_banks:
                    if k[1]==saver_interlocutore.banca_di_appartenenza:
                      for l in saver.Trust_banks:
                        if l[1]==k[1]:
                          Tb1=l[0]   # <---------------- inizializzata solo nell'if, non fuori
                        else:
                          Tb1=0
                      Tb=(Tb1+k[0]*T)/(1+T)
                      Tb=float(f"{Tb:.3f}")
                      b=[Tb,k[1],ciclo] #<---------------- questa è la trust verso una banca nella lista delle trust del saver: [trust_banca, banca, ciclo]
                      agg_or_sos2(saver.Trust_banks, b)
                      #print(f"Trust_banks saver {saver.unique_id} = {saver.Trust_banks}")
                  if saver_interlocutore.info_var_equity!=[]:
                    c=random.choice(saver_interlocutore.info_var_equity)
                    agg_or_sos2(saver.info_var_equity,c)
                    for k, bank in enumerate(saver.Trust_banks):
                      if bank[1]==c[1]:
                        t=saver.Trust_banks[k][0]
                        t=0.5*(t+sigmoid(c[0]))
                        t=float(f"{t:.3f}")
                        saver.Trust_banks[k][0]=t
                  if saver_interlocutore.info_var_clienti!=[]:
                    d=random.choice(saver_interlocutore.info_var_clienti)
                    agg_or_sos2(saver.info_var_clienti,d)
                    for z, bank in enumerate(saver.Trust_banks):
                      if bank[1]==d[1]:
                        q=saver.Trust_banks[z][0]
                        q=0.5*(q+sigmoid(d[0]))
                        q=float(f"{q:.3f}")
                        saver.Trust_banks[z][0]=q



             # print(f"Trust saver {saver.unique_id} = {saver.Trust}")
              #print(f"Trust_banks saver {saver.unique_id} = {saver.Trust_banks}")
          print(f"\n")

          for i, saver in enumerate(self.schedule_savers.agents):       #saver smette di parlare
              saver.marker=-1





         # I Savers cambiano Banca

          for i, bank in enumerate(self.schedule_banks.agents):
            bank.bank_old_numero_clienti = bank.numero_clienti # qui salvo anche il numero di clienti vecchio


          for saver in self.schedule_savers.agents:
            #print(f"saver{saver.unique_id} ha banca= {saver.banca_di_appartenenza}")
            for trust in saver.Trust_banks:
              if trust[1]==saver.banca_di_appartenenza:
                if trust[0]<0.35:
                  if trust[0]!=max(saver.Trust_banks)[0]:
                    nuova_trust= max(saver.Trust_banks)
                    for bank in self.schedule_banks.agents:
                      if nuova_trust[1]==bank.unique_id:
                        saver.banca_di_appartenenza= nuova_trust[1]
                        print(f"saver{saver.unique_id} ho cambiato  con banca= {saver.banca_di_appartenenza}")















          # Quanti savers ci sono in ciascuna banca?

          for i, bank in enumerate(self.schedule_banks.agents):
            numero=0
            for j, saver in enumerate(self.schedule_savers.agents):
              if saver.banca_di_appartenenza==bank.unique_id:
                numero+=1
            bank.numero_clienti=numero

          for bank in self.schedule_banks.agents:
            print(f"Bank {bank.unique_id} numero_clienti= {bank.numero_clienti}")
            print(f"Bank {bank.unique_id} old_numero_clienti= {bank.bank_old_numero_clienti}")
            bank.variation_numero_clienti= bank.numero_clienti-bank.bank_old_numero_clienti
            print(f"Bank {bank.unique_id} variation_numero_clienti= {bank.variation_numero_clienti}")
            info2=[bank.variation_numero_clienti, bank.unique_id,ciclo]
            self.blog.receive_info(info2)
          print(f"\n")


          # cambia i depositi della banca (non tiene conto delle flutuazioni?)
          print(f"\n")
          for i, bank in enumerate(self.schedule_banks.agents):
            bank.cash_depositi=0
            for j, saver in enumerate(self.schedule_savers.agents):
              if saver.banca_di_appartenenza==bank.unique_id:
                bank.cash_depositi += saver.bank_account
           # print(f"Bank {bank.unique_id} cash_depositi= {bank.cash_depositi}")
          print(f"\n")


          #guardo la liquidità--> sotto soglia? mercato interbancario
          for bank in self.schedule_banks.agents:
            bank.liquidity=0
            bank.liquidity= bank.cash_depositi-bank.bank_loans
            bank.disponibilita = bank.liquidity - 0.1*bank.cash_depositi
            if bank.liquidity<0.1*bank.cash_depositi:
              print(f"Bank {bank.unique_id} è in rosso")
              bank.bank_request=0.1*bank.cash_depositi-bank.liquidity
          print(f"\n")


          for bank_loan in self.schedule_banks.agents:
            list_interest_banks= []
            bank_loan.list_vero= []
            if bank_loan.bank_request!=0:
              m=0
              n=0
              while bank_loan.bank_request>0:
                print(f"Bank {bank_loan.unique_id} bank_request= {bank_loan.bank_request}")
                n+=1
                for bank in self.schedule_banks.agents:
                  bank.vero = 0
                  if bank.disponibilita-bank_loan.bank_request > 0:
                    bank.vero = bank_loan.bank_request
                  else:
                    bank.vero = bank.disponibilita

                  if bank.vero > 0 and bank.bank_loans + bank.vero < bank.bank_equity*10 and bank.bank_loans + bank.vero < bank.cash_depositi*0.9:
                    interest_bank=0.04+random.uniform(-3,3)/100
                    interest_bank=float(f"{interest_bank:.3f}")
                    list_interest_banks.append([interest_bank, bank.unique_id])
                    m+=1

                if list_interest_banks!=[] and len(list_interest_banks)>1: ###############
                    bank_interest=0
                    bank_interest=min(list_interest_banks)  # vincitore dell'asta
                    print(f"Interesse scelto: {bank_interest}")
                    for bank in self.schedule_banks.agents:
                      if bank_interest[1]==bank.unique_id:
                          bank_loan.list_vero.append([bank.vero,bank_interest[0],bank.unique_id])
                          bank.bank_loans += bank.vero
                          print(f"Quantità a disposizione = {bank.disponibilita}, quantità prestata = {bank.vero} dalla banca {bank.unique_id}")
                          bank.disponibilita -= bank.vero
                          bank_loan.bank_request -= bank.vero
                          print(f"Bank_request= {bank_loan.bank_request} della banca {bank_loan.unique_id}")
                          print(f"Bank_request= {sum(sublist[0] for sublist in bank_loan.list_vero)} della banca {bank_loan.unique_id}")
                          if bank_loan.bank_request ==0:
                            print("pimpa")
                            break

                if list_interest_banks == []:
                  break
                if n>m:
                  break

# COSE DA FARE
# restituzione del prestito


          #loans rida soldi
          for i, loan in enumerate(self.schedule_loans.agents):
            x=random.uniform(0,100)
            if loan.loan_interest != None:

                #print(f"loan {loan.unique_id} è stato approvato")
                for bank in self.schedule_banks.agents:
                  if loan.loan_interest[1]==bank.unique_id:
                    if loan.prob_default<x:
                      bank.bank_equity += loan.loan_request * loan.loan_interest[0]
                    else:
                    # print(f"loan {loan.unique_id} è stato rifiutato")
                      bank.bank_equity -= loan.loan_request


            #banca da interessi a saver                           ######### problemino da risolvere: se la banca ha perdite a causa dei loans e va a equity sotto zero non può ridare interessi perche viene rimossa, noi la rimozione ce l abbiamo dopo il pagamento interessi e non va bene
          for bank in self.schedule_banks.agents:
            for saver in self.schedule_savers.agents:
              if saver.banca_di_appartenenza==bank.unique_id:
                interest=0
                interest = saver.bank_account*0.01
                saver.bank_account+= interest
                bank.bank_equity-= interest


          # Banca restituisce prestiti mercato interbancario

          for bank_loan in self.schedule_banks.agents:
            if bank_loan.list_vero!=[] and bank_loan.bank_equity>sum(sublist[0] for sublist in bank_loan.list_vero):
                for bank in self.schedule_banks.agents:
                  for sublist in bank_loan.list_vero:
                    if sublist[2]==bank.unique_id:
                      bank_loan.bank_equity -= sublist[0]*(1+sublist[1])
                      bank.bank_equity += sublist[0]*sublist[1]
            elif bank_loan.list_vero!=[] and bank_loan.bank_equity<sum(sublist[0] for sublist in bank_loan.list_vero):
                for bank in self.schedule_banks.agents:
                  for sublist in bank_loan.list_vero:
                    if sublist[2]==bank.unique_id:
                      if bank_loan.bank_equity-sublist[0]*(1+sublist[1])>0:
                        bank_loan.bank_equity-=sublist[0]*(1+sublist[1])
                        bank.bank_equity+=sublist[0]*sublist[1]

                      else:
                        for bc in self.schedule_central_bank.agents:
                         self.blog.send_info_bc(bank_loan,bc)
                         a = bc.default(bank_loan)
                         if a == 0:
                          bank.bank_equity -= sublist[0]
                         else:
                          bank.bank_equity += sublist[0]*(1+sublist[1])




#banca calcola variazione equity
          for bank in self.schedule_banks.agents:
            bank.bank_loans=0
            #print(f"Bank {bank.unique_id} equity= {bank.bank_equity}")
            #print(f"Bank {bank.unique_id} old_equity= {bank.old_equity}")
            bank.variation_equity= bank.bank_equity-bank.old_equity
            bank.variation_equity=float(f"{bank.variation_equity:.3f}")
            info=[bank.variation_equity, bank.unique_id,ciclo]
            self.blog.receive_info(info)
            print(f"Bank {bank.unique_id} variation_equity= {bank.variation_equity}")
            if bank.bank_equity<=0:
              for bc in self.schedule_central_bank.agents:
                self.blog.send_info_bc(bank,bc)
                bc.default(bank)


          for bank in self.schedule_banks.agents:
            if bank.bank_default==1:
              self.banks_to_remove.append(bank)


      #saver cambiano banca
          for regione in self.regioni:
           for bank in self.banks_to_remove:
              if bank in regione.banks_per_regione:
                  regione.banks_per_regione.remove(bank)
                  print(f"Agente {bank.unique_id} rimosso dalla regione")
                  available_banks = [b for b in regione.banks_per_regione if b.bank_default == 0]
                  # Trova nuovi banchi per i clienti della banca in default
                  savers_to_update = [saver for saver in self.schedule_savers.agents if saver.banca_di_appartenenza == bank.unique_id]
                  for saver in savers_to_update:
                      if available_banks!=[]:
                          new_bank = random.choice(available_banks)
                          saver.banca_di_appartenenza = new_bank.unique_id
                          print(f"Saver {saver.unique_id} riassegnato alla banca {new_bank.unique_id}")
                      else:
                          print(f"Errore: Nessuna banca valida disponibile per Saver {saver.unique_id} nella regione")





          # Controllare gli agenti rimanenti nel scheduler
          print("Agenti rimanenti nel scheduler:")
          for agent in self.schedule_banks.agents:
            print(agent.unique_id)

          print("Agenti rimanenti nel banks_per_regione:")
          for regione in self.regioni:
            for bank in regione.banks_per_regione:
              print(bank.unique_id)


          for regioni in self.regioni:
            for bank in regioni.banks_per_regione:
              print(f"bank{bank.unique_id} ha regione= {bank.bank_regione}")
              print(f"bank{bank.unique_id} ha {bank.bank_default}")



         # for saver in self.schedule_savers.agents:
            #print(f"saver{saver.unique_id} ha banca= {saver.banca_di_appartenenza}")
          print(f"\n")

          self.blog.send_info_onniscente()
          self.blog.send_info_local()


          for saver in self.schedule_savers.agents:
            for trust in saver.Trust_banks:
              for bank in self.banks_to_remove:
                if bank.unique_id==trust[1]:
                  saver.Trust_banks.remove(trust)
            #print(saver.Trust_banks)


          for loan in self.schedule_loans.agents:
            loan.loan_interest=[]

          #for saver in self.schedule_savers.agents:
           #print(f"saver {saver.unique_id} banca di appartenenza {saver.banca_di_appartenenza}")

          for saver in self.schedule_savers.agents:
            for bank in self.banks_to_remove:
              if saver.banca_di_appartenenza==bank.unique_id:
                print(f"saver{saver.unique_id} in banca {bank.unique_id} problematica")

      def remove_bank_agent(self, agent):
          self.schedule_banks.remove(agent)
          print(f"Agente {agent.unique_id} rimosso dal scheduler")

      def clean_up_default_banks(self):
          agents_to_remove = [bank for bank in self.schedule_banks.agents if bank.bank_default == 1]

          for bank in agents_to_remove:
              self.remove_bank_agent(bank)
              for regione in self.regioni:
                  if bank in regione.banks_per_regione:
                      #regione.banks_per_regione.remove(bank)
                      print(f"Agente {bank.unique_id} rimosso dalla regione {regione}")

model = Network(300,100,20,3)  # numero_savers loan banche regioni


for i in range(20):
    print(f"ciclo {i}")
    model.blog.numero_ciclo(i)
    model.clean_up_default_banks()
    model.step(0.1,i)
    print(f"\n")


```
